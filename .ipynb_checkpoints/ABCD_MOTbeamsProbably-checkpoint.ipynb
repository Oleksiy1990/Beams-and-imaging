```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline

from pyConstants import *
from SrConstants import *
```

Define the functions for ABCD matrices and for finding and aanlyzing the Gaussian beam q-parameter

```python
def lens_matrix(f):
    return np.matrix([[1,0],[-1/f,1]])
def space_matrix(dist,n):
    return np.matrix([[1,dist/n],[0,1]])
def q_param(lam,waist,rad):
    q = 1/(1/rad - lam*1j/(np.pi*waist**2))
    return q
def propagate_q(q_in,matr):
    q_out = (matr[0,0]*q_init+matr[0,1])/(matr[1,0]*q_init+matr[1,1])
    return q_out
def rayleigh_r(q):
    inv = 1/q
    rayl = -1/inv.imag
    return rayl
def waist_from_q(q,lam):
    inv = 1/q
    rayl = -1/inv.imag
    w = np.sqrt(lam*rayl/np.pi)
    return w
```

The basic test to see that the code works and produces expected results

```python
# we take all units in mm
f1 = 30
f2 = 150
dist_between = 180 

wavelength = 461e-6 #mm
init_beamdiam = 1


lenM1 = lens_matrix(f1)
lenM2 = lens_matrix(f2) #we expect factor 5 magnification 
matr_in_between = space_matrix(dist_between,1)

lengths_after = np.linspace(1,1000) # up to 1 m

matr_after = [space_matrix(length,1) for length in lengths_after]

sys_matrix = [matr_a*lenM2*matr_in_between*lenM1 for matr_a in matr_after]

q_init = q_param(wavelength,init_beamdiam,np.inf) #in mm
q_final = [propagate_q(q_init,sys_matr) for sys_matr in sys_matrix]
waists = [waist_from_q(q_fin,wavelength) for q_fin in q_final]

plt.plot(lengths_after,waists)
plt.title("Lenses: f1 = %.i, f2 = %.i, dist = %.i"%(f1,f2, dist_between))
plt.xlabel("Position after 2nd lens [mm]")
plt.ylabel("Beam waist diam [mm]")
plt.show()

f1 = -30
f2 = 150
dist_between = 120 

wavelength = 461e-6 #mm
init_beamdiam = 1


lenM1 = lens_matrix(f1)
lenM2 = lens_matrix(f2) #we expect factor 5 magnification 
matr_in_between = space_matrix(dist_between,1)

lengths_after = np.linspace(1,1000) # up to 1 m

matr_after = [space_matrix(length,1) for length in lengths_after]

sys_matrix = [matr_a*lenM2*matr_in_between*lenM1 for matr_a in matr_after]

q_init = q_param(wavelength,init_beamdiam,np.inf) #in mm
q_final = [propagate_q(q_init,sys_matr) for sys_matr in sys_matrix]
waists = [waist_from_q(q_fin,wavelength) for q_fin in q_final]

plt.plot(lengths_after,waists)
plt.title("Lenses: f1 = %.i, f2 = %.i, dist = %.i"%(f1,f2, dist_between))
plt.xlabel("Position after 2nd lens [mm]")
plt.ylabel("Beam waist diam [mm]")
plt.show()


# we're supposed to see the 5 mm collimated beam in both cases
```

##### Now we want to test the real thing for our intended achromatic doublets

First we assume both positive lenses, one is f = 19, the other one is f = 150 (somehow we don't have an f=20 achromat for 
half-inch)

```python
# we take all units in mm
# assume that blue is ideally positioned and we look at the results for red
# we calculate the errors by taking error(689) - error (461) from the datasheet on Thorlabs  


f1 = 19 + 0.0946 # this is the focal length difference between blue and red
f2 = 150 + 0.231 # this is the focal length difference between blue and red
dist_between_list = 169 + np.linspace(-1,1,5) # we introduce errors, assuming that the spacing between the lenses could be 
                                              # plus or minus 1 mm from ideal 

wavelength = 461e-6 #mm
init_beamdiam = 3.6

for dist_between in dist_between_list:
    lenM1 = lens_matrix(f1)
    lenM2 = lens_matrix(f2) #we expect factor 5 magnification 
    matr_in_between = space_matrix(dist_between,1)

    lengths_after = np.linspace(1,1000) # up to 1 m

    matr_after = [space_matrix(length,1) for length in lengths_after]

    sys_matrix = [matr_a*lenM2*matr_in_between*lenM1 for matr_a in matr_after]

    q_init = q_param(wavelength,init_beamdiam,np.inf) #in mm
    q_final = [propagate_q(q_init,sys_matr) for sys_matr in sys_matrix]
    waists = [waist_from_q(q_fin,wavelength) for q_fin in q_final]
    plt.plot(lengths_after,waists)

plt.legend(np.linspace(-1,1,5))
plt.title("Lenses: f1 = 19, f2 = 150, dist = 169")
plt.xlabel("Position after 2nd lens [mm]")
plt.ylabel("Beam waist diam [mm]")
plt.show()

# we expect the beam to be collimated with w ~ 28 mm 
# The different lines show the different misalignments as given in the legend 
```

Now we assume one positive, one negative lens, one is f = -20, the other one is f = 150
